```python
import networkx as nx
import matplotlib.pyplot as plt

# Insert an edge to the graph
def add_edge(graph, source, target):
    graph.add_edge(source, target)

# Remove an edge from the graph
def remove_edge(graph, source, target):
    if graph.has_edge(source, target):
        graph.remove_edge(source, target)

# Visualize the graph
def visualize_graph(graph, title="Graph Visualization"):
    pos = nx.spring_layout(graph)  # Spring layout for a visually pleasing arrangement
    plt.figure(figsize=(10, 8))
    nx.draw(graph, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray')
    plt.title(title)
    plt.show()




```

```python
    G = nx.karate_club_graph()
    print("Graph loaded successfully!")

    # Display the initial graph
    visualize_graph(G, "Original Zachary Karate Club Graph")

    # Perform some operations
    add_edge(G, 1, 34)  # Add a new edge between two nodes (1 and 34)
    remove_edge(G, 1, 2)  # Remove an existing edge between two nodes (1 and 2)

    # Display the modified graph
    visualize_graph(G, "Modified Graph with Added and Removed Edges")

    # Perform some graph operations
    print("Number of nodes:", G.number_of_nodes())
    print("Number of edges:", G.number_of_edges())
    print("Is the graph connected?", nx.is_connected(G))
```
