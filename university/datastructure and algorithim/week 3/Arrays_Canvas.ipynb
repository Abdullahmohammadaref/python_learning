# **Array Operations:**



*   Insertion at beginning






```python
import numpy as np
```

```python
def insert_at_beginning(array, element):
    # Create a new array with an extra space for the new element
    new_array = np.zeros(len(array) + 1, dtype=array.dtype)

    # Insert the new element at the beginning
    new_array[0] = element

    # Copy the original elements to the new array, shifted by one position
    for i in range(len(array)):
        new_array[i + 1] = array[i]

    return new_array


```

```python
original_array = np.array([1, 2, 3, 4])
element = 10

updated_array = insert_at_beginning(original_array, element)
print("Updated Array:", updated_array)
```

```python
original_array = np.array([1, 2, 3, 4])
element = 10
location = 4

updated_array = np.insert(original_array, location, element)
print(updated_array)
```

Inserting new element at the specified position

```python
def insert_into_array(array, element, position):
    if position < 0 or position > len(array):
        print("Position is out of bounds")
        return array  # Return the original array if position is invalid

    # Create a new NumPy array with an extra space for the new element
    new_array = np.zeros(len(array) + 1, dtype=array.dtype)

    # Copy elements before the insertion position
    for i in range(position):
        new_array[i] = array[i]

    # Insert the new element at the specified position
    new_array[position] = element

    # Copy the rest of the elements after the inserted element
    for i in range(position, len(array)):
        new_array[i + 1] = array[i]

    return new_array
```

```python
original_array = np.array([1, 2, 3, 4])
new_element = 99
position = 2

result = insert_into_array(original_array, new_element, position)
print(result)
```

Deleting the element at the beginning

```python
def delete_from_beginning(array):
    # If the array is empty, return an empty array
    if len(array) == 0:
        return array

    # Create a new array with one less space
    new_array = np.zeros(len(array) + 1, dtype=array.dtype)

    # Copy elements from the original array, starting from the second element
    for i in range(1, len(array)):
        new_array[i - 1] = array[i]

    return new_array
```

```python
array =  np.array([1, 2, 3, 4])

updated_array = delete_from_beginning(array)
print("Updated Array:", updated_array)
```

```python
array =  np.array([1, 2, 3, 4])
position = 2

updated_array = np.delete(array, position)
print(updated_array)
```

Deleting the element at the specified position

```python
def delete_at_position(array, position):
    # Check if position is within bounds
    if position < 0 or position >= len(array):
        print("Position is out of bounds")
        return array  # Return the original array if the position is invalid

    # Create a new array with one less space
    new_array = np.zeros(len(array) + 1, dtype=array.dtype)

    # Copy elements before the specified position
    for i in range(position):
        new_array[i] = array[i]

    # Copy elements after the specified position, shifted one position to the left
    for i in range(position + 1, len(array)):
        new_array[i - 1] = array[i]

    return new_array
```

```python
array =  np.array([1, 2, 3, 4])
position = 2

updated_array = delete_at_position(array, position)
print("Updated Array:", updated_array)
```

# Blurring the Image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the image using OpenCV
image = cv2.imread('/content/drive/MyDrive/Gisma/Data_structure/image_processing/lenna.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.show()
```

```python
# Define a 7x7 averaging kernel
kernel = np.ones((7, 7), dtype=np.float32) / 49  # Each element is 1/49 to average the 7x7 region

# Get the dimensions of the image
rows, cols = image.shape

# Create an empty array to store the blurred image
blurred_image = np.zeros_like(image, dtype=np.float32)
#np.zeros_like creates a new array of zeros with the same shape and type as an existing array.

# Apply the 7x7 kernel to each pixel in the image (excluding the edges)
for i in range(3, rows - 3):
    for j in range(3, cols - 3):
        # Extract the 7x7 region around the current pixel
        region = image[i - 3:i + 4, j - 3:j + 4]

        # Apply the kernel (element-wise multiplication and summation)
        blurred_image[i, j] = np.sum(region * kernel)

# Convert the blurred image back to an 8-bit format
blurred_image = np.clip(blurred_image, 0, 255).astype(np.uint8)

# Display the original and blurred images
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title("Blurred Image (7x7 Kernel)")
plt.axis("off")

plt.show()
```

# Adding a Border

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('/content/drive/MyDrive/Gisma/Data_structure/image_processing/gray.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()

# Define border size and color
border_size = 20  # Number of pixels for the border
border_color = 128  # Gray color for the border (0 to 255 for grayscale; (0, 0, 0) to (255, 255, 255) for color images)


# Get the dimensions of the original image
original_height, original_width = image.shape

# Create a new array with additional space for the border
new_height = original_height + 2 * border_size
new_width = original_width + 2 * border_size
bordered_image = np.full((new_height, new_width), border_color, dtype=np.uint8)
#np.full creates a new array filled with a specified value, rather than zeros or ones
#np.uint8 stands for "unsigned 8-bit integer," which means each value is an integer between 0 and 255.

# Place the original image in the center of the new array
bordered_image[border_size:border_size + original_height, border_size:border_size + original_width] = image

# Display the image with border
plt.imshow(bordered_image, cmap='gray')
plt.title("Image with Border")
plt.axis("off")
plt.show()
```
